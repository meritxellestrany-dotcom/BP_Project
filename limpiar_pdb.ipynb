**Pymol commands to clean the structure**

```python
import pymol
pymol.finish_launching(['pymol', '-cq'])  # headless, quiet

from pymol import cmd

# Load structure
cmd.load("data/6M0J.pdb")

```

```python
# To remove other chains
cmd.remove("not chain A+E")
```

```python
# To remove heteroatoms and water
cmd.remove("hetatm")
cmd.remove("resn HOH")
```

```python
# AFEGIR: fix alternate locations
cmd.remove("alt B")
cmd.alter("all", "alt=''")
```

```python
# Save the cleaned structure
cmd.save("data/6m0j_clean_bo.pdb")
cmd.quit()
```

```python
import biobb_structure_checking
import biobb_structure_checking.constants as cts
from biobb_structure_checking.structure_checking import StructureChecking
#from biobb_vs.pdbqt_prepare import PdbqtPrepare
base_dir_path=biobb_structure_checking.__path__[0]
args = cts.set_defaults(base_dir_path,{'notebook':True}) #if working with jupyter notebook put True
```

```python
#APLICAR 6m0j_check tutorial per structure checking
base_dir = "data/"
args['input_structure_path'] = base_dir+'6m0j_clean_bo.pdb'
args['output_structure_path'] = base_dir+'6m0j_fixed_bo.pdb'
args['output_structure_path_charges'] = base_dir+'6m0j_fixed_bo.pdbqt'
args['debug'] = False
args['verbose'] = False
#Added missing defaults
args['time_limit'] = 3600
args['nocache'] = False
args['copy_input'] = None
args['build_warnings'] = False
args['coords_only'] = False
args['output_format'] = 'pdbqt'

#Initializing checking engine, loading structure and showing statistics
st_c = StructureChecking(base_dir_path, args)
print(st_c.chains())
```

```python
st_c.chiral('All')
```

```python
st_c.backbone('--fix_atoms All --fix_chain none --add_caps none')
```

```python
st_c.fixside('All')
```

```python
opts = {
    'add_mode': 'auto',
    'add_charges': 'CMIP'
}

st_c.add_hydrogen(opts)
```

```python
st_c.checkall()
```

```python
st_c._save_structure(args['output_structure_path_charges'])
```

```python
args['output_format'] = 'pdb'
st_c._save_structure(args['output_structure_path'])
```
